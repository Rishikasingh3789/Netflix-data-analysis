```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C://Users//singh//Downloads//netflix.title.zip")
df
```

```python
df.head()
```

```python
df.isnull().sum()

# Missing values handle
df = df.dropna(subset=['country', 'rating'])
df
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Growth Over Years")
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.show()
```

```python
## Key Insights

- Movies dominate over TV Shows on Netflix  
- United States produces the highest amount of content  
- Content growth increased rapidly after 2015  
- Drama and Comedy are the most popular genres  
- Netflix follows a diverse content strategy
```
